```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import poisson

lambda_ = 8  # rata-rata pelanggan per jam 
hours = 12  # jumlah jam operasional per hari 
days = 30  # jumlah hari simulasi 
total_hours = hours * days


simulated_data = poisson.rvs(mu=lambda_, size=total_hours)


plt.figure(figsize=(10, 6)) 
plt.hist(simulated_data, bins=range(0, max(simulated_data)+2), alpha=0.75, color='skyblue', edgecolor='black', density=True) 
plt.title('Histogram Distribusi Kedatangan Pelanggan per Jam') 
plt.xlabel('Jumlah Pelanggan per Jam') 
plt.ylabel('Frekuensi Relatif') 
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show()

prob_no_customers = poisson.pmf(0, lambda_)

prob_more_than_10 = 1 - poisson.cdf(10, lambda_)

print(f"Probabilitas tidak ada pelanggan dalam 1 jam: {prob_no_customers:.5f}") 
print(f"Probabilitas lebih dari 10 pelanggan dalam 1 jam: {prob_more_than_10:.5f}")

x_values = range(0, max(simulated_data)+1) 
theoretical_probabilities = poisson.pmf(x_values, lambda_)

plt.figure(figsize=(12, 6)) 
plt.hist(simulated_data, bins=range(0, max(simulated_data)+2), alpha=0.5, color='skyblue', edgecolor='black', density=True, label='Simulasi') 
plt.plot(x_values, theoretical_probabilities, marker='o', linestyle='-', color='red', label='Teoretis Poisson') 
plt.title('Perbandingan Distribusi Simulasi dan Teoretis Poisson') 
plt.xlabel('Jumlah Pelanggan per Jam') 
plt.ylabel('Frekuensi Relatif / Probabilitas') 
plt.legend() 
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show()
```
